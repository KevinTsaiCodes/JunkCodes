### Activation Function 激励函数

激勵函數都是 非線性函數，而普通神經網路出來的數據，一般都是線性的

## Import Library

```python
import torch
from torch.nn import functional as F # 引入函數
from torch.autograd import Variable # 反向回饋
import matplotlib.pyplot as plt
```

### Create fake data

#### x values

```python
x = torch.linspace(-5,5,200) # linspace method: 把所有線段分成一點一點的，linspace(起始值, 終值, 中間分成幾個點)
x = Variable(x)
x_np = x.data.numpy() # 由於 torch 為 tensor 型態，所以須將它轉為 numpy 的型態，這樣 matplotlib 才可識別
```

#### y values

```python
y_relu = F.relu(x).data.numpy()
y_sigmoid = F.sigmoid(x).data.numpy()
y_tanh = F.tanh(x).data.numpy()
y_softplus = F.softplus(x).data.numpy() # softmax 也是激勵函數，但是它用於概率比較多
```

### Visualize Activation Function

```python
plt.figure(1, figsize=(8, 6))
plt.subplot(221)
plt.plot(x_np, y_relu, c='red', label='relu')
plt.ylim((-1, 5))
plt.legend(loc='best')

plt.subplot(222)
plt.plot(x_np, y_sigmoid, c='blue', label='sigmoid')
plt.ylim((-0.2, 1.2))
plt.legend(loc='best')

plt.subplot(223)
plt.plot(x_np, y_tanh, c='green', label='tanh')
plt.ylim((-1.2, 1.2))
plt.legend(loc='best')

plt.subplot(224)
plt.plot(x_np, y_softplus, c='orange', label='softplus')
plt.ylim((-0.2, 6))
plt.legend(loc='best')

plt.show()
```
